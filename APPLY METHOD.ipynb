```python
#apply() method is used to broadcast a function to columns
```

```python
#lets read a df first
```

```python
import pandas as pd
```

```python
df=pd.read_csv("Iris.csv")
```

```python
df
```

```python
#Create a new column that stores length of species name
```

```python
df["lenspec_name"]=df["Species"].apply(len)
```

```python
df
```

```python
#Create a function that takes a number and returns number multiplied by 3
```

```python
#using apply method, create a new column that contains the doubled sepal
#length
```

```python
def times2(num):
    return num*2
```

```python
df["sepallength2"]=df["SepalLengthCm"].apply(times2)
```
